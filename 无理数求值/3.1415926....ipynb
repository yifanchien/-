```python
from __future__ import division
def pi_calculate(number,save=False,show=True,la=1000):
    import time
    import sys
    if show:
        time1 = time.time()
        b = 10**(number+10)
        x1 = b*4//5
        x2 = b//-239
        he = x1+x2
        b = he
        for i in range(3,number*2,2):
            x1 //= -25
            x2 //= -57121
            x = (x1+x2)//i
            he += x
            if b == he:
                break
            b = he
            j = int(((i-1)//2)/(number-1)*la)*"█"
            jx = (la - int(((i-1)//2)/(number-1)*la))*"░"
            sys.stdout.write(f"\r第{(i-1)//2}次 共{number-1}次 {j}{jx}    {str(he*4//10**10)[0]}.{str(he*4//10**10)[1:]}")
            sys.stdout.flush()
        if save:
            with open("pi.txt","w") as f:
                f.write(f"{str(he*4//10**10)[0]}.{str(he*4//10**10)[1:]}")
        print(f"    time:{time.time()-time1}s")
    else:
        time1 = time.time()
        b = 10**(number+10)
        x1 = b*4//5
        x2 = b//-239
        he = x1+x2
        b = he
        for i in range(3,number*2,2):
            x1 //= -25
            x2 //= -57121
            x = (x1+x2)//i
            he += x
            if b == he:
                break
            b = he
        if save:
            with open("pi.txt","w") as f:
                f.write(f"{str(he*4//10**10)[0]}.{str(he*4//10**10)[1:]}")
        print(f"{str(he*4//10**10)[0]}.{str(he*4//10**10)[1:]}\ntime:{time.time()-time1}s")
    return he*4//10**10/(10*number)
```

```python
pi = pi_calculate(5000)
```
