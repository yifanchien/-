```python
%matplotlib notebook
from sympy import *
import matplotlib.pyplot as plt
def draw2D(function, scope, float1):
    fig = plt.figure(figsize = ((8,6)))
    ax = fig.add_subplot(1,1,1)
    xs = []
    ys = []
    n = "2"
    for x in range(-scope, scope+1):
        x *= float1
        if x != 0:
            ys.append(eval(function))
            xs.append(x)
    ax.scatter(xs, ys, label='function1')
    ax.legend()
    str4 = input("Draw a again? (n)/y")
    while str4 == "y":
        function = input("Please input function.")
        scope = int(input("Please enter the range of X.(A number)"))
        xs = []
        ys = []
        n = "2"
        for x in range(-scope, scope+1):
            x *= float1
            if x != 0:
                ys.append(eval(function))
                xs.append(x)
        ax.scatter(xs, ys, label='function'+n)
        ax.legend()
        n = str(int(n)+1)
        str4 = input("Draw a again? (n)/y")
```

```python
%matplotlib notebook
def draw3D(function, scope1, scope2, float1):
    from mpl_toolkits.mplot3d import Axes3D
    fig = plt.figure(figsize = ((8,6)))
    ax = fig.add_subplot(1,1,1,projection='3d')
    xs = []
    ys = []
    zs = []
    n = "2"
    y = -scope2
    for x in range(-scope1, scope1+1):
        for y in range(-scope2, scope2+1):
            x *= float1
            y *= float1
            zs.append(eval(function))
            xs.append(x)
            ys.append(y)
        y+=1
    ax.scatter(xs, ys, zs, label='function1')
    ax.legend()
    str3 = input("Draw a again? (n)/y")
    while str3 == "y":
        function = input("Please input function.")
        scope1 = int(input("Please enter the range of X.(A number)"))
        scope2 = int(input("Please enter the range of Y.(A number)"))
        xs = []
        ys = []
        zs = []
        y = -scope2
        for x in range(-scope1, scope1+1):
            for y in range(-scope2, scope2+1):
                x *= float1
                y *= float1
                zs.append(eval(function))
                xs.append(x)
                ys.append(y)
        ax.scatter(xs, ys, zs, label='function'+n)
        ax.legend()
        n = str(int(n)+1)
        str3 = input("Draw a again? (n)/y")
```

```python
str1 = input("Please input the dimensions of the function.(2,3):")
if str1 == "2":
    function = input("Please input function.")
    scope = int(input("Please enter the range of X.(A number)"))
    draw2D(function, scope, float(input("Please input function precision (A decimal)")))
elif str1 == "3":
    function = input("Please input function.")
    scope1 = int(input("Please enter the range of X.(A number)"))
    scope2 = int(input("Please enter the range of Y.(A number)"))
    draw3D(function, scope1, scope2, 1)
else:
    print("Dimension is too high (low) can't display.")
str2 = input("Drawing again? (n)/y")
while str2 == "y":
    str1 = input("Please input the dimensions of the function.(2,3):")
    if str1 == "2":
        function = input("Please input function.")
        scope = int(input("Please enter the range of X.(A number)"))
        draw2D(function, scope, float(input("Please input function precision (A decimal)")))
    elif str1 == "3":
        function = input("Please input function.")
        scope1 = int(input("Please enter the range of X.(A number)"))
        scope2 = int(input("Please enter the range of Y.(A number)"))
        draw3D(function, scope1, scope2, 1)
    else:
        print("Dimension is too high (low) can't display.")
    str2 = input("Drawing again? (n)/y")
```
