```python
import random
import numpy as np
import matplotlib.pyplot as plt
EX_NUMBER = 1e3
BIGBOOK_FIGHT = {}
DIEBOOK = []
J_LIST = []
R_LIST = []
S_LIST = []
ALL_LIST = []
LEN = []
```

```python
def create_ex(EX_NUMBER):
    SPACE = {}
    for i in range(EX_NUMBER):
        X,Y,C,CLASS = str(hex(random.randint(0,4095)))[2:],str(hex(random.randint(0,4095)))[2:],str(hex(random.randint(1,15)))[2:],random.choices(["J","R","S"])[0]
        if CLASS == "J":
            AVAILABLE_RESOURCES_SLOPE = random.randint(1,10)
            AVAILABLE_RESOURCES = int(eval("0x"+C))*AVAILABLE_RESOURCES_SLOPE*random.randint(10,15)
            PREDATORY_RATE_SLOPE = 11-AVAILABLE_RESOURCES_SLOPE
            PREDATORY_RATE = int(eval("0x"+C))*PREDATORY_RATE_SLOPE*random.randint(10,15)
        elif CLASS == "R":
            AVAILABLE_RESOURCES_SLOPE = random.randint(1,10)
            AVAILABLE_RESOURCES = int(eval("0x"+C))*AVAILABLE_RESOURCES_SLOPE*random.randint(10,15)
            PREDATORY_RATE_SLOPE = 11-AVAILABLE_RESOURCES_SLOPE
            PREDATORY_RATE = int(eval("0x"+C))*PREDATORY_RATE_SLOPE*random.randint(8,12)
        else:
            AVAILABLE_RESOURCES_SLOPE = random.randint(1,10)
            AVAILABLE_RESOURCES =int(eval("0x"+C))*AVAILABLE_RESOURCES_SLOPE*random.randint(4,10)
            PREDATORY_RATE_SLOPE = 11-AVAILABLE_RESOURCES_SLOPE
            PREDATORY_RATE = int(eval("0x"+C))*PREDATORY_RATE_SLOPE*random.randint(1,5)
        if len(X) == 3:
            if len(Y) == 3:
                SPACE[f"{CLASS+X+Y}"] = [C,AVAILABLE_RESOURCES,PREDATORY_RATE]
            elif len(Y) == 2:
                SPACE[f"{CLASS+X+'0'+Y}"] = [C,AVAILABLE_RESOURCES,PREDATORY_RATE]
            else:
                SPACE[f"{CLASS+X+'00'+Y}"] = [C,AVAILABLE_RESOURCES,PREDATORY_RATE]
        elif len(X) == 2:
            SPACE[f"{CLASS+'0'+X+Y}"] = [C,AVAILABLE_RESOURCES,PREDATORY_RATE]
        else:
            SPACE[f"{CLASS+'00'+X+Y}"] = [C,AVAILABLE_RESOURCES,PREDATORY_RATE]
        print(f"\r{i/EX_NUMBER}",end="")
    with open("space_ex.txt","w") as f:
        f.write(str(SPACE))
def name_to_information(name):
    return name[0],int(eval("0x"+name[1:4])),int(eval("0x"+name[4:]))
def draw_points(SPACE):
    global LEN
    S_X_list,S_Y_list = [],[]
    R_X_list,R_Y_list = [],[]
    J_X_list,J_Y_list = [],[]
    DFE = {}
    JI,RI,SI = 0,0,0
    for i in SPACE.keys():
        if name_to_information(i)[0] == "S":
            S_X_list.append(name_to_information(i)[1])
            S_Y_list.append(name_to_information(i)[2])
            SI += 1
        elif name_to_information(i)[0] == "R":
            R_X_list.append(name_to_information(i)[1])
            R_Y_list.append(name_to_information(i)[2])
            RI += 1
        else:
            J_X_list.append(name_to_information(i)[1])
            J_Y_list.append(name_to_information(i)[2])
            JI += 1
    print(f"""
-------------------------------------------------------
J:{JI}                 R:{RI}                   S:{SI}
-------------------------------------------------------""")
    for i in SPACE.keys():
        DFE[SPACE[i][0]] = 0
    print(DFE.keys())
    LEN.append(len(DFE))
    plt.scatter(S_X_list,S_Y_list)
    plt.scatter(R_X_list,R_Y_list)
    plt.scatter(J_X_list,J_Y_list)
    plt.show()
    J_LIST.append(JI)
    R_LIST.append(RI)
    S_LIST.append(SI)
    ALL_LIST.append(JI+RI+SI)
def change(SPACE):
    i0 = 0
    for i in SPACE.keys():
        if i[0] == "J":
            SPACE[i] = [str(hex(int(eval("0x"+SPACE[i][0]))+random.randint(0,5)))[2:],SPACE[i][1],SPACE[i][2]]
            C = SPACE[i][0]
            AVAILABLE_RESOURCES_SLOPE = random.randint(1,10)
            AVAILABLE_RESOURCES = int(eval("0x"+C))*AVAILABLE_RESOURCES_SLOPE*random.randint(10,15)/10
            PREDATORY_RATE_SLOPE = 11-AVAILABLE_RESOURCES_SLOPE
            PREDATORY_RATE = int(eval("0x"+C))*PREDATORY_RATE_SLOPE*random.randint(10,15)/10
            SPACE[i] = [str(hex(int(eval("0x"+SPACE[i][0]))))[2:],SPACE[i][1]+AVAILABLE_RESOURCES,SPACE[i][2]+PREDATORY_RATE]
        elif i[0] == "R":
            SPACE[i] = [str(hex(int(eval("0x"+SPACE[i][0]))+random.randint(1,2)))[2:],SPACE[i][1],SPACE[i][2]]
            C = SPACE[i][0]
            AVAILABLE_RESOURCES_SLOPE = random.randint(1,10)
            AVAILABLE_RESOURCES = int(eval("0x"+C))*AVAILABLE_RESOURCES_SLOPE*random.randint(10,15)/10
            PREDATORY_RATE_SLOPE = 11-AVAILABLE_RESOURCES_SLOPE
            PREDATORY_RATE = int(eval("0x"+C))*PREDATORY_RATE_SLOPE*random.randint(4,10)/10
            SPACE[i] = [str(hex(int(eval("0x"+SPACE[i][0]))))[2:],SPACE[i][1]+AVAILABLE_RESOURCES,SPACE[i][2]+PREDATORY_RATE]
        else:
            SPACE[i] = [str(hex(int(eval("0x"+SPACE[i][0]))+random.randint(0,1)))[2:],SPACE[i][1],SPACE[i][2]]
            C = SPACE[i][0]
            AVAILABLE_RESOURCES_SLOPE = random.randint(1,10)
            AVAILABLE_RESOURCES = int(eval("0x"+C))*AVAILABLE_RESOURCES_SLOPE*random.randint(8,12)/10
            PREDATORY_RATE_SLOPE = 11-AVAILABLE_RESOURCES_SLOPE
            PREDATORY_RATE = int(eval("0x"+C))*PREDATORY_RATE_SLOPE*random.randint(4,10)/10
            SPACE[i] = [str(hex(int(eval("0x"+SPACE[i][0]))))[2:],SPACE[i][1]+AVAILABLE_RESOURCES,SPACE[i][2]+PREDATORY_RATE]
        i0 += 1
        print(f"\r{i0/len(SPACE)}",end="")
    return SPACE
def change_cut(SPACE):
    EX_NAME1 = random.choice(list(SPACE.keys()))
    EX_NAME2 = random.choice(list(SPACE.keys()))
    if EX_NAME1 != EX_NAME2:
        EX_DF1,EX_DF2 = int(eval("0x"+SPACE[EX_NAME1][0])),int(eval("0x"+SPACE[EX_NAME2][0]))
        nm,nq = SPACE[EX_NAME1],SPACE[EX_NAME2]
        a,b = SPACE[EX_NAME2][2]*(EX_DF2/(EX_DF1+EX_DF2))*2,SPACE[EX_NAME1][2]*(EX_DF1/(EX_DF1+EX_DF2))*2
        SPACE[EX_NAME1] = [SPACE[EX_NAME1][0],SPACE[EX_NAME1][1]-a+b*random.randint(3,10)/10,SPACE[EX_NAME1][2]]
        SPACE[EX_NAME2] = [SPACE[EX_NAME2][0],SPACE[EX_NAME2][1]-b+a*random.randint(3,10)/10,SPACE[EX_NAME2][2]]
        print(f"""-------------------------------------------
        {EX_NAME1}:{nm}
        {EX_NAME2}:{nq}
        {EX_NAME1} <====> {EX_NAME2}
        {EX_NAME1}:{SPACE[EX_NAME1]}
        {EX_NAME2}:{SPACE[EX_NAME2]}
-------------------------------------------""")
        if (EX_NAME1 in BIGBOOK_FIGHT) and (EX_NAME2 in BIGBOOK_FIGHT):
            BIGBOOK_FIGHT[EX_NAME1] = BIGBOOK_FIGHT[EX_NAME1]+[EX_NAME2]
            BIGBOOK_FIGHT[EX_NAME2] = BIGBOOK_FIGHT[EX_NAME2]+[EX_NAME1]
        elif EX_NAME1 in BIGBOOK_FIGHT:
            BIGBOOK_FIGHT[EX_NAME1] = BIGBOOK_FIGHT[EX_NAME1]+[EX_NAME2]
            BIGBOOK_FIGHT[EX_NAME2] = [EX_NAME1]
        elif EX_NAME2 in BIGBOOK_FIGHT:
            BIGBOOK_FIGHT[EX_NAME1] = [EX_NAME2]
            BIGBOOK_FIGHT[EX_NAME2] = BIGBOOK_FIGHT[EX_NAME2]+[EX_NAME1]
        else:
            BIGBOOK_FIGHT[EX_NAME1] = [EX_NAME2]
            BIGBOOK_FIGHT[EX_NAME2] = [EX_NAME1]
    return SPACE
def cut(SPACE):
    print("-------------------------------------------")
    SPACE_NEW = {}
    for i in SPACE.keys():
        if SPACE[i][1] > 0:
            SPACE_NEW[i] = SPACE[i]
        else:
            print(i,"灭亡",SPACE[i])
            DIEBOOK.append(i)
    SPACE = SPACE_NEW
    print(f"剩余 {len(SPACE)}")
    print("-------------------------------------------")
    return len(SPACE),SPACE
def save(EX_N_LIST,SPACE_D):
    with open("BIGBOOK_FIGHT.txt","w") as f1:
        f1.write(str(BIGBOOK_FIGHT))
    with open("DIEBOOK.txt","w") as f2:
        f2.write(str(DIEBOOK))
    with open("ALL_LIST.txt","w") as f3:
        f3.write(str([J_LIST,R_LIST,S_LIST,ALL_LIST]))
    with open("EX_N_LIST.txt","w") as f4:
        f4.write(str(EX_N_LIST))
    with open("SPACE_D.txt","w") as f5:
        f5.write(str(SPACE_D))
    with open("LEN.txt","w") as f6:
        f6.write(str(LEN))
```

```python
def main(EX_NUMBER,show=True,wait=False,save_=True,EX_N_MAX=1e100):
    create_ex(int(EX_NUMBER))
    print("\n")
    with open("space_ex.txt","r") as f:
        SPACE = eval(f.read())
    SPACE_D = SPACE
    import time
    EX_N = EX_N_MAX
    EX_N_LIST = []
    while EX_N>1:
        for i in range(10):
            SPACE = change_cut(SPACE)
        EX_N,SPACE = cut(SPACE)
        EX_N_LIST.append(EX_N)
        SPACE = change(SPACE)
        if show:draw_points(SPACE)
        if wait:time.sleep(wait)
    if save_:save(EX_N_LIST,SPACE_D)
```

```python
main(1e5)
```
